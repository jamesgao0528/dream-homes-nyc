```python
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, recall_score, roc_auc_score
```

```python
dat_reg = pd.read_csv(r"C:\Users\User\Downloads\dat_reg.csv")
dat_class = pd.read_csv(r"C:\Users\User\Downloads\dat_class.csv")
```

```python
5.
```

```python
import numpy as np

# Q5: RMSE for pred1 (用 numpy 手动算)
mse_pred1 = mean_squared_error(dat_reg['sales'], dat_reg['pred1'])
rmse_pred1 = np.sqrt(mse_pred1)
print("Q5 - RMSE for pred1:", rmse_pred1)
```

```python
6.
```

```python
r2_pred1 = r2_score(dat_reg['sales'], dat_reg['pred1'])
print("Q6 R-squared for pred1:", r2_pred1)
```

```python
7.
```

```python
mse_pred1 = mean_squared_error(dat_reg['sales'], dat_reg['pred1'])
mse_pred2 = mean_squared_error(dat_reg['sales'], dat_reg['pred2'])
mse_pred3 = mean_squared_error(dat_reg['sales'], dat_reg['pred3'])

print("Q7 - MSE for pred1:", mse_pred1)
print("Q7 - MSE for pred2:", mse_pred2)
print("Q7 - MSE for pred3:", mse_pred3)
lowest_mse_model = min(
    {'pred1': mse_pred1, 'pred2': mse_pred2, 'pred3': mse_pred3},
    key=lambda k: {'pred1': mse_pred1, 'pred2': mse_pred2, 'pred3': mse_pred3}[k]
)
print("Q7 Lowest MSE Model:", lowest_mse_model)
```

```python
8.
```

```python
y_true = dat_class['sold']
y_pred1 = (dat_class['pred1'] > 0.5).astype(int)

accuracy_pred1 = accuracy_score(y_true, y_pred1)
print("Q8 Accuracy for pred1:", accuracy_pred1)
```

```python
9.
```

```python
y_pred2 = (dat_class['pred2'] > 0.5).astype(int)

sensitivity_pred2 = recall_score(y_true, y_pred2)
print("Q9 Sensitivity for pred2:", sensitivity_pred2)
```

```python
10.
```

```python
auc_pred1 = roc_auc_score(y_true, dat_class['pred1'])
auc_pred2 = roc_auc_score(y_true, dat_class['pred2'])
auc_pred3 = roc_auc_score(y_true, dat_class['pred3'])

print("Q10 - AUC for pred1:", auc_pred1)
print("Q10 - AUC for pred2:", auc_pred2)
print("Q10 - AUC for pred3:", auc_pred3)
highest_auc_model = max(
    {'pred1': auc_pred1, 'pred2': auc_pred2, 'pred3': auc_pred3},
    key=lambda k: {'pred1': auc_pred1, 'pred2': auc_pred2, 'pred3': auc_pred3}[k]
)
print("Q10 Highest AUC Model:", highest_auc_model)
```
