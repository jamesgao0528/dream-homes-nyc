```python
import pandas as pd

jan = pd.read_csv(r"C:\Users\User\Desktop\january.csv")
apr = pd.read_csv(r"C:\Users\User\Desktop\april.csv")
jul = pd.read_csv(r"C:\Users\User\Desktop\july.csv")
oct = pd.read_csv(r"C:\Users\User\Desktop\october.csv")

df = pd.concat([jan, apr, jul, oct], ignore_index=True)

df.to_csv(r"C:\Users\User\Desktop\t100_combined.csv", index=False)

print(df.head())
```

```python
df = pd.read_csv(r"C:\Users\User\Desktop\t100_combined.csv")

df = df[df["PASSENGERS"] > 10]

df.to_csv(r"C:\Users\User\Desktop\t100_passenger_only.csv", index=False)

print(df.head())
```

```python
airport_flow = df.groupby("ORIGIN")["PASSENGERS"].sum().reset_index()
print(airport_flow)
```

```python
airport_flow.to_csv(r"C:\Users\User\Desktop\airport_passenger_flow.csv", index=False)
```

```python
df = pd.read_csv(r"C:\Users\User\Desktop\t100_passenger_only.csv")

airport_month_passengers = df.groupby(["MONTH","ORIGIN"])["PASSENGERS"].sum().reset_index()

airport_month_passengers.to_csv(r"C:\Users\User\Desktop\airport_month_passengers.csv", index=False)

print(airport_month_passengers.head())
```
